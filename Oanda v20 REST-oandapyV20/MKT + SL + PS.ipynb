```python
import pandas as pd
import oandapyV20
import oandapyV20.endpoints.orders as orders
```

```python
accountID = ''
access_token = ''
```

```python
client = oandapyV20.API(access_token=access_token)
r = orders.OrderList(accountID)
client.request(r)
store = []
# Check for current open orders
for oo in r.response['orders']:
    store.append(oo)
pd.DataFrame(store)
```

[Order Definitions](http://developer.oanda.com/rest-live-v20/order-df/#collapse_definition_2)

```python
data = \
{
  "order": {
        "type": "MARKET",
        "units": "260",
        "instrument": "GBP_USD",
        "timeInForce": "FOK",
        "takeProfitOnFill": {
            "timeInForce": "GTC",
            "price": "1.35"
        },
        "stopLossOnFill": {
            "timeInForce": "GTC",
            "price": "1.30"
    }
  }
}
```

```python
r = orders.OrderCreate(accountID, data=data)
client.request(r)
```

```python
client = oandapyV20.API(access_token=access_token)
r = orders.OrderList(accountID)
client.request(r)
store = []
# Check for current open orders
for oo in r.response['orders']:
    store.append(oo)
```
