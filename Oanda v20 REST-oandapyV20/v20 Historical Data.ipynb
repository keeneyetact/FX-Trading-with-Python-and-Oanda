```python
import pandas as pd
import oandapyV20
import oandapyV20.endpoints.instruments as instruments
```

```python
accountID = ''
access_token = ''
```

```python
client = oandapyV20.API(access_token=access_token)
```

http://developer.oanda.com/rest-live-v20/instrument-df/#CandlestickGranularity

```python
params = {
          "count": 250,
          "granularity": "D"}
```

```python
r = instruments.InstrumentsCandles(instrument="EUR_USD",
                                   params=params)
```

```python
client.request(r)
```

```python
r.response['candles'][0]['mid']
```

```python
r.response['candles'][0]['time']
```

```python
r.response['candles'][0]['volume']
```

```python
dat = []
for oo in r.response['candles']:
    dat.append([oo['time'], oo['volume'], oo['mid']['o'], oo['mid']['h'], oo['mid']['l'], oo['mid']['c']])
```

```python
df = pd.DataFrame(dat)
df.columns = ['Time', 'Volume', 'Open', 'High', 'Low', 'Close']
df = df.set_index('Time')
df.head()
```
